```python
# List of column indices based on the features you want to select for PCA
columns_to_keep = [4, 5, 7, 8, 9, 10, 11, 12, 13]  # Adjust based on your dataset structure

# Select the relevant columns from the dataset
data = df.iloc[:, columns_to_keep]

# Rename the columns for clarity
data.columns = [
    'xCO2_SW_wet', 'xCO2_Air_wet', 'pCO2_SW_sat', 'pCO2_Air_sat', 'dfCO2', 'dpCO2', 'SST_C', 'Salinity', 'pH_total_scale'
]

# Display the first few rows of the selected data
print(data.head())
```

```python
print(data.dtypes)  # Check data types of all columns
```

```python
# Select only numeric columns
data_numeric = data.select_dtypes(include=['number'])

# Standardize the numeric data
data_standardized = (data_numeric - data_numeric.mean()) / data_numeric.std()

print("\nStandardized Data Preview:")
print(data_standardized.head())
```

```python
# Step 5: Visualize correlations in standardized data
import seaborn as sns
import matplotlib.pyplot as plt

print("\nCorrelation Heatmap:")
corr_matrix = data_standardized.corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")  
plt.title("Correlation Heatmap of Standardized Features")
plt.show()
```

```python
# Step 6: Compute the covariance matrix
cov_matrix = data_standardized.cov()
print("\nCovariance Matrix:")
cov_matrix
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)  # Reduce to 2 components
principal_components = pca.fit_transform(data_standardized)

# Create a DataFrame with the PCA results
pca_df = pd.DataFrame(principal_components, columns=['PC1', 'PC2'])

# Show the explained variance of each principal component
print(f'Explained variance by PC1: {pca.explained_variance_ratio_[0]:.2f}')
print(f'Explained variance by PC2: {pca.explained_variance_ratio_[1]:.2f}')
```

```python
# Visualize the PCA results
plt.figure(figsize=(8, 6))
sns.scatterplot(x='PC1', y='PC2', data=pca_df)
plt.title('PCA of Ocean Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
# Step 7: Calculate eigenvalues and eigenvectors
import numpy as np
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
```

```python
# Sort eigenvectors by eigenvalues in descending order
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]
```

```python
# Step 8: Transform the data into the new subspace
principal_components = data_standardized.dot(eigenvectors)
```

```python
# Step 9: Print and interpret results
print("\nEigenvalues:")
print(pd.Series(eigenvalues, index=[f'PC{i+1}' for i in range(len(eigenvalues))]))
```

```python
print("\nPrincipal Components:")
principal_components
```

```python
explained_variance_ratio = eigenvalues / sum(eigenvalues)
cumulative_explained_variance = np.cumsum(explained_variance_ratio)
print("\nExplained Variance Ratio:")
print(pd.Series(explained_variance_ratio, index=[f'PC{i+1}' for i in range(len(explained_variance_ratio))]))

print("\nCumulative Explained Variance:")
print(pd.Series(cumulative_explained_variance, index=[f'PC{i+1}' for i in range(len(cumulative_explained_variance))]))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Example data (Replace this with your data)
data = np.array([
    [2.5, 2.4, 3.5],
    [0.5, 0.7, 1.2],
    [2.2, 2.9, 3.0],
    [1.9, 2.2, 2.8],
    [3.1, 3.0, 3.9],
    [2.3, 2.7, 3.1],
    [2, 1.6, 2.5],
    [1, 1.1, 1.4],
    [1.5, 1.6, 2.0],
    [1.1, 0.9, 1.2]
])

# Standardizing the data
scaler = StandardScaler()
data_standardized = scaler.fit_transform(data)

# Apply PCA
pca = PCA(n_components=2)
pca.fit(data_standardized)

# Extract the loadings (coefficients of each variable in the principal components)
loadings = pca.components_.T  # Transpose to match features with components

# Plot the correlation circle
plt.figure(figsize=(8, 8))

# Plot the arrows (loadings)
for i in range(loadings.shape[0]):
    plt.arrow(0, 0, loadings[i, 0], loadings[i, 1], 
              color='b', alpha=0.7, head_width=0.05, head_length=0.1)
    plt.text(loadings[i, 0] * 1.15, loadings[i, 1] * 1.15, f'Feature {i+1}', color='b', ha='center', va='center')

# Set the aspect ratio to be equal
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.axhline(0, color='k', lw=1)
plt.axvline(0, color='k', lw=1)

# Add grid and title
plt.grid(True)
plt.title('Correlation Circle Plot')

# Display the plot
plt.show()
```
